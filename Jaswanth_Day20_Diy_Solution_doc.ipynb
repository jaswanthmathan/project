Q1. Problem Statement: Measures of Dispersion
Load the “HRdataset_v14.csv” dataset into a DataFrame and group the data
based on the “Department” column. From the grouped DataFrame and perform
the following tasks:
1. Calculate the Mean and Median, of the “Salary” column.
2. Find out which Department has the highest number of employees
3. Calculate the standard deviation and variance of the “Salary” column
4. Find the interquartile range of the "Salary" column
5. Find skewness and kurtosis of the original dataset
6. Find the z-score of the "Salary" column
7. Find the 30th percentile of the "Salary" column
8. Find the Quartiles of the "GenderID" column (You can set quartile values of
your choice)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
hr = pd.read_csv('HRDataset_v14.csv')
hr
```

```python
#Analyzing the data.
hr.info()
```

```python
#Grouping data based on "Department" column.
grouped_hr =  hr.groupby(['Department'])
print((grouped_hr).first())
```

```python
#Average Salary(Mean)
dept_Salary_hr =  hr.groupby(['Department'],as_index=False).Salary.mean()
display(dept_Salary_hr)
```

```python
#Calculating median of salary column
dept_Salary_hr =  hr.groupby(['Department'],as_index=False).Salary.median()
display(dept_Salary_hr)
```

```python
#Calculating the standard deviation and variance of the "Salary" column.
#Standard deviation of salary column
print("Department-wise Standard deviation of salary:______")
dept_Salary_hr['Salary'].std()
```

```python
print("Department-wise Variance of salary:______")
dept_Salary_hr['Salary'].var()
```

```python
#Calculation which department has highest employees
print("This department has highest number of employees")
display(hr['Department'].mode())
```

```python
# Find interquartile range of "Salary" column
from scipy.stats import iqr
print("Interquartile range of salary:______")
iqr(dept_Salary_hr['Salary'])
```

```python
#Find skewness and kurtosis of the original dataset
print("Skewness in Original Dataset: __________")
print(hr.skew())
```

```python
print("Kurtosis in Original Dataset: __________")
print(hr.kurt())
```

```python
#Finding z-score of "Salary" column.
from scipy.stats import zscore
zscores = zscore(dept_Salary_hr['Salary'])
print("Department-wise z-score of salaries:____________ ")
display(zscores)
```

```python
import numpy
ptl = numpy.percentile(hr["Salary"], 30)
print("30th percentile of Salary column is:")
print(ptl)
```

```python
#Finding Quartiles of "GenderID" column. 
x = numpy.quantile([20, 40], hr["GenderID"])
print("Quartile values of 20, 40 from GenderID column:")
print(x)
```
