Write a Python program with two DataFrames, with columns “name” and “age”.
Now, combine both the DataFrames vertically using the concat() function,
without the duplicate rows, and print the final DataFrame.


```python
import pandas as pd
#create df1

df1 = pd.DataFrame({  'name': ['shubham','adin','rahul'],    'age': [23,28,30]})
display(df1)
```

```python
# create df2 
df2 = pd.DataFrame({  'name': ['amrit','shubham','ali'],    'age': [20,23,28]})
display(df2)
```

```python
#final dataframe
final_df = pd.concat([df1, df2],ignore_index=True).drop_duplicates()

d=final_df.reset_index(drop=True)

# final DataFrame
display(final_df)
```

Write a Python program to create two DataFrames. First DataFrame with
columns “employee” and “department,” Second DataFrame with columns
“employee” and “joining_date,” and Perform the following tasks:
1. Merge the DataFrames based on the “employee” column
2. Display the final DataFrame with employees records who are working in the
“Engineering” department

```python
#import pandas as pd

# create dfa
dfa = pd.DataFrame({"employee": ["Bob", "jake", "lisa", "sue","john","Daisy"], 
                    "department": ["Accounting","Engineering", "Engineering","HR", "Engineering", "Engineering"]})
display(dfa)

```

```python
#create dfb

dfb = pd.DataFrame({  "employee": ["lisa", "Bob", "jake", "sue","john","Daisy"],
   "joining_date": ["2004", "2008", "2012", "2022","2021","2020"]
})
display(dfb)
```

```python
merged_df = pd.merge(dfa, dfb, on="employee")
display(merged_df)
```

```python
engineering_df = merged_df[merged_df["department"] == "Engineering"]
display(engineering_df)
```
