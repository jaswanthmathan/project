Load the “cereal.csv” data into a DataFrame and perform the following tasks:
1. Explore the DataFrame using info() and describe() functions
2. Replace the manufacturer names with the dictionary of names given here –
{'A':'American Home Food Products', 'G':'General Mills',
'K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}
3. Find out the top five cereal manufacturers based on customer ratings
4. With the help of a barplot, compare the manufacturers based on customer
ratings
5. With the help of a pie chart, find out which of the manufacturers are the
largest producers of cereals


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
cereals = pd.read_csv('cereal.csv')
cereals
```

```python

print(cereals.info())
```

```python

print(cereals.describe())
```

```python
manufacturer_dict = {'A': 'American Home Food Products', 'G': 'General Mills', 'K': 'Kelloggs', 
                     'N': 'Nabisco', 'P': 'Post', 'Q': 'Quaker Oats', 'R': 'Ralston Purina'}
cereals['mfr'] = cereals['mfr'].replace(manufacturer_dict)
display (cereals['mfr'])
```

```python
display(cereals)
```

```python
top_manufacturers = cereals.sort_values(by='rating',ascending=False).reset_index(drop=True).head()
display(top_manufacturers)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='mfr', y='rating', data=cereals)
plt.xticks(rotation=45)
plt.show()
```

```python
manufacturer_counts = cereals['mfr'].value_counts()
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%')
plt.show()
```
